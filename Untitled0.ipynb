```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
link = 'http://ireland.cx/teaching/1920/svwork/ii/infotheory/autocorrelation.txt'
data = pd.read_csv(link, header=None)
```

```
arr = data.to_numpy()
arr = arr.flatten()
```

```
plt.plot(arr[:100])
plt.show()
```

```
autocorrelation = np.correlate(arr,arr,'same')
autocorrelation[:5]
```

```
plt.plot(autocorrelation[:100])
plt.show()
```

```
autocorrelation.argmax() # Always returns 0

inflection = np.diff(np.sign(np.diff(autocorrelation))) # Find the second-order differences
peaks = (inflection < 0).nonzero()[0] + 1 # Find where they are negative
delay = peaks[autocorrelation[peaks].argmax()] # Of those, find the index with the maximum value
delay
```
